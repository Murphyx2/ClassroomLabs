```python
# importing numpy
import numpy as np
```

```python
# Create a numpy array with the elements "1, 2, 3, 5, 0", print the array and test if all of its elements are non-zero.
array1 = np.array([1,2,3,5,0])
print("Elements of the Array are: ", array1)
print("are all the elements in the array non-zero?: ", array1.all())
```

```python
# Create a numpy array with the elements "3, -1, 3, 7, 2", print the array and test if any of its elements is non-zero.
array2 = np.array([3,-1,3,7,2])
print("Elements of the Array are: ", array2)
print("Is any the elements in the array a non-zero?: ", array2.any())
```

```python
# Create a numpy array with the elements "0, 0, 0, 0", print the array and test if any of its elements is non-zero.
array3 = np.array([0,0,0,0])
print("Elements of the Array are: ", array3)
print("Is any the elements in the array a non-zero?: ", array3.all())
```

```python
# Create a numpy array with the elements "56, -2, 7", print the array and the amount of memory required to store it.
array4 = np.array([56,-2,7])
print("Elements of the Array are: ", array4)
print("Amount of memory required for this array: ", array4.itemsize*array4.size, "bits") 
```

```python
# Create a numpy array with 10 zeros and print the array.
array_zeros = np.zeros(10, dtype=int)
print("Elements of the Array are: ", array_zeros)
# Create a numpy array with 6 ones and print the array.
array_ones = np.ones(6, dtype=int)
print("Elements of the Array are: ", array_ones)
# Create a numpy array with 12 '3's and print the array. [3 3 3 3 3 3 3 3 3 3 3 3]
array_threes = 3*np.ones(12, dtype=int)
print("Elements of the Array are: ", array_threes)
```

```python
# Create a numpy array of the integers from 8 to 20 and print the array. (make sure 20 itself is included) hint:  use "arange" method
array5 = np.arange(8,21,1, dtype=int)
print("Elements of the Array are: ", array5)

# Create a numpy array of the even integers from 8 to 20 and print the array. (make sure 20 itself is included)   hint:  use "arange" method
array6 = np.arange(8,21,2, dtype=int)
print("Elements of the Array are: ", array6)
```

```python
# Create a numpy array consisting of 100 random elements from normal distribution with mean of 2 and std of 0.5. Print the mean, max, min and std of the array.
random_array = np.random.normal(2, 0.5, 100)
print("Size of the array: ", random_array.size)
print("Mean of the array: ", random_array.mean())
print("Max of the array: ", random_array.max())
print("Min of the array: ", random_array.min())
print("Std of the array: ", random_array.std())
print("Elements of the Array are: ", random_array)
# Create a numpy array consisting of 1000 random elements from normal distribution with mean of 2 and std of 0.5. Print the mean, max, min and std of the array.
random_array2 = np.random.normal(2, 0.5, 1000)
print("Size of the array: ", random_array2.size)
print("Mean of the array: ", random_array2.mean())
print("Max of the array: ", random_array2.max())
print("Min of the array: ", random_array2.min())
print("Std of the array: ", random_array2.std())
print("Elements of the Array are: ", random_array2)
# Create a numpy array consisting of 1000000 random elements from normal distribution with mean of 2 and std of 0.5. Print the mean, max, min and std of the array.
random_array3 = np.random.normal(2, 0.5, 10000)
print("Size of the array: ", random_array3.size)
print("Mean of the array: ", random_array3.mean())
print("Max of the array: ", random_array3.max())
print("Min of the array: ", random_array3.min())
print("Std of the array: ", random_array3.std())
print("Elements of the Array are: ", random_array3)
# (Bonus question) Looking at the three above examples with 1000, 5000, and 10000 random elements, do you have any observations about mean and std values?
#I noticed that the increase in element improve the precision of the mean and std values. The more elements the closer to the element we use in the setup of the array. 
```
